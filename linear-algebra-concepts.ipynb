### 1. Representing data as linear systems

#### What are linear systems and why are they so important to machine learning?

Linear systems represent set of entities that model a **straight line relationship** between variables like, $y = mx + b$ where **x,y** are variables which can have no solution or unique solution or ∞ solutions and **m is the slope** (the rate of change of y with respect to x) and **b is the y-intercept** (where y crosses the y axis i.e. when x=0).

When a system has no solution, it is called a contradictory system. e.g. $x + y = 10$ and $x + y = 20$.
When a system has ∞ solutions, it is called a redundant system. This happen when the # unique variables > # unique equations. e.g.  $x + y = 10$ and  $2x + 2y = 20$.
When a system has a unique solution, it is called a complete system. e.g. $2x + y = 7$ and $x + y = 5$.

Anything which is not in the given format ($y = mx + b$), like quadratic equations, log equations, trignometric equations are all non-linear equations.

Linear equations are useful in math to understand linear relationship between n variables. In ML, we can easily frame prediction problems using linear equations assuming the dependent variable as a resultant of linear relationship between the dependent variables.




```python
# numpy array is often used for linear algebra

import numpy as np
import matplotlib.pyplot as plt
m = np.array([2, 1])
b = np.array([7, 5])
x = np.linspace(-10, 10, 200) # to generate 200 numbers between -10 and 10
y1 = m[0]*x + b[0] # 2x + 7 = y
y2 = m[1]*x + b[1] # x + 5 = y
plt.figure(figsize=(4, 3))
plt.plot(x, y1,)
plt.plot(x, y2,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear System with 2 variables in 2D space')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.show()
```

```python
# Let's plot for the 3 systems of Lin Alg
m2 = np.array([2, 1])
b2 = np.array([4, 2])
y3 = (m2[0]*x + b2[0])/2 # 2x + 4 = 2y
y4 = m2[1]*x + b2[1] # x + 2 = y

m3 = np.array([3, 3])
b3 = np.array([7, 2])
y5 = m3[0]*x + b3[0] # 3x + 7 = y
y6 = m3[1]*x + b3[1] # 3x + 2 = y


plt.figure(figsize=(15, 4))

# --- Graph 1 ---
plt.subplot(1, 3, 1)  
plt.plot(x, y1,color = 'orange')
plt.plot(x, y2,color = 'teal')
plt.title('Complete System')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)

# --- Graph 2 ---
plt.subplot(1, 3, 2)
plt.plot(x, y3, color = 'orange')
plt.plot(x, y4, color = 'teal')
plt.title('Redundant System')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)

# --- Graph 3 ---
plt.subplot(1, 3, 3)
plt.plot(x, y5,color = 'orange')
plt.plot(x, y6,color = 'teal')
plt.title('Contradictory System')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)

# Show all plots
plt.tight_layout()
plt.show()
```
